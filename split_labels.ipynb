## Read csv and concat together

```python
import pandas as pd
df_crop = pd.read_csv('./data/crop_download_test.csv')
df_s3 = pd.read_csv('./data/s3upload_download_test.csv')
```

```python
df_crop.head()
```

```python
df_s3.head()
```

```python
new_df = pd.concat([df_crop, df_s3], ignore_index=True)
new_df = pd.read_csv('./data/20191106_tight_boxed_data.csv')
```

```python
# drop duplicate with origin url

new_df = new_df.drop_duplicates(subset=['origin_url'], keep='last')
```

```python
new_df = new_df.reset_index()
```

```python
new_df = new_df.drop(['index'],axis=1)
```

```python
new_df
```

## remove url part of filename

```python
for i, row in new_df.iterrows():
    new_df.at[i,'origin_url'] = new_df['origin_url'][i].split('/')[-1]
```

## arrange pandas dataframe form

```python
new_df['class'] = 'handbag'
```

```python
new_df = new_df[['origin_url', 'width', 'height', 'class', 'left', 'top', 'right', 'bottom']]
```

```python
new_df = new_df.rename(columns={'origin_url' : 'filename', 'left' : 'xmin', 'right' : 'xmax', 'top': 'ymin', 'bottom' : 'ymax'})
```

```python
new_df = new_df[:1500]
```

## split training / test data and write to csv

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(new_df, test_size = 0.25)
```

```python
df_train
```

```python
df_train.to_csv('./data/training_bag_csv_no_handle_v2', index=False)
df_test.to_csv('./data/test_bag_csv_no_handle_v2', index=False)
```
